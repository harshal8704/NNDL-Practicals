```python
import numpy as np
```

```python
#Vector Addition
v1 = np.array([20,40,60])
v2 = np.array([10,20,30])
```

```python
print("Vector 1 = ",v1)
print("Vector 2 = ",v2)
```

```python
result = v1 + v2
```

```python
print("Vector Addition =",result)
```

```python
#Matrix Multiplication

m1 = np.array([[2,4,6],[1,2,3]])
m2 = np.array([[2,4],[1,2],[4,6]])
```

```python
print("Matrix 1 = ",m1)
print("Matrix 2 = ",m2)
```

```python
matrixresult = np.matmul(m1,m2)
```

```python
print("Matrix Multiplication = ",matrixresult)
```
